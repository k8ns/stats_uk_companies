```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import chart_studio.plotly as py
import cufflinks as cf
import seaborn as sns
import plotly.express as px
%matplotlib inline

# Make Plotly work in your Jupyter Notebook
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
# Use Plotly locally
cf.go_offline()
```

```python
companies = pd.read_csv('companies.csv')
companies = companies.sort_values('Year')
companies.head()
```

### Top 20 cities with the most registered number of companies

```python
plt.figure(figsize=(18,7))
sns.countplot(y='City', data=companies, orient="h", order = companies['City'].value_counts().iloc[:20].index)
```

### Registered companies per year

```python
plt.figure(figsize=(18,10))
sns.countplot(y='Year', data=companies, orient="h")
```

### Top 50 activities

```python
plt.figure(figsize=(18,20))
sns.countplot(y='SicDescription', data=companies, orient="h", order = companies['SicDescription'].value_counts().iloc[:50].index)
```

### Top 50 activities registered in 2020

```python
plt.figure(figsize=(18,20))
companies2020 = companies[companies['Year'] == 2020]
companies2020
sns.countplot(y='SicDescription', data=companies2020, orient="h", order = companies2020['SicDescription'].value_counts().iloc[:50].index)
```
